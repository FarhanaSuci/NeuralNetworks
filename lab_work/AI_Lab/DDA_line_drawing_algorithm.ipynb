```python
#B190305001
import matplotlib.pyplot as B190305001
def DDA_Line_Algorithm(x0, y0, x1, y1):
    
    #difference of x
    dx = abs(x0 - x1)
    #difference of y
    dy = abs(y0 - y1)
    
    
    steps = max(dx, dy)
    X_coordinate_steps = dx / steps
    Y_coordinate_steps = dy / steps
    x = float(x0)
    y = float(y0)

    x_coordinates = []
    y_coordinates = []
    for i in range(steps):
        
        x_coordinates.append(x)
        y_coordinates.append(y)

        x = x + X_coordinate_steps
        y = y + Y_coordinate_steps

    B190305001.plot(x_coordinates, y_coordinates, marker="o", markersize=1, markerfacecolor="green")
    B190305001.show()

    return list(zip(x_coordinates, y_coordinates))

#main_Code
x0, y0 = 15, 15
x1, y1 = 70, 80
DDA_points = DDA_Line_Algorithm(x0, y0, x1, y1)
print("The Intermediary points are:", DDA_points)
```
