**Write a program to implement the DDA line algorithm**

```python
from matplotlib import pyplot as suchi 

def DDA_Line_Draw(x0, y0, x1, y1): 
  
    dx = abs(x0 - x1) 
    dy = abs(y0 - y1) 
    steps = max(dx, dy) 
    x_increment = dx/steps 
    y_increment = dy/steps 
    x = float(x0) 
    y = float(y0) 
    x_coorinates = [] 
    y_coorinates = [] 
  
    for i in range(steps): 
        x_coorinates.append(x) 
        y_coorinates.append(y) 
        x = x + x_increment 
        y = y + y_increment 
  
    suchi.plot(x_coorinates, y_coorinates, marker="o", 
             markersize=2, markerfacecolor="blue") 
    suchi.show() 
  
if __name__ == "__main__": 
  
   
    x0, y0 = 30, 30
    x1, y1 = 70,80
    DDA_Line_Draw(x0, y0, x1, y1) 
```

**Write a program to implement midpoint line drawing algorithm**

```python
def midPoint_Line_Draw(X1,Y1,X2,Y2):  
    dx = X2 - X1  
    dy = Y2 - Y1  
    d = dy - (dx/2)  
    x = X1 
    y = Y1   
    print(x,",",y,"\n") 
    while (x < X2): 
        x=x+1 
        if(d < 0): 
            d = d + dy   
        else: 
            d = d + (dy - dx)  
            y=y+1
      
        suchi.plot(x,y, marker="o", 
        markersize=5, markerfacecolor="red") 
         
        print(x,",",y,"\n")  
     
  
if __name__=='__main__': 
    X1 = 4
    Y1 = 4
    X2 = 12
    Y2 = 7
    suchi.plot(X1,Y1, marker="o", 
    markersize=5, markerfacecolor="red") 
    suchi.plot(X2,Y2, marker="o", 
    markersize=5, markerfacecolor="red") 
    midPoint_Line_Draw(X1, Y1, X2, Y2)  
```
